```python
%load_ext autoreload
%autoreload 2

from amber.store import LocalStore
from amber.adapters.text_snippet_dataset import TextSnippetDataset

import torch

from amber.adapters.language_model import LanguageModel

model = LanguageModel.from_huggingface(
    "sshleifer/tiny-gpt2",
)
dataset = TextSnippetDataset.from_huggingface(
    "roneneldan/TinyStories",
    split="train",
    cache_dir="./cache/tinystories",
    text_field="text",
    limit=100
)

store = LocalStore("./activations_store")
```

```python
model.get_layer_names()
```

```python
from amber.activations import save_model_activations

layer_signature = 'gpt2lmheadmodel_lm_head'

save_model_activations(
    model,
    dataset,
    store,
    'test_run',
    layer_signature,
    batch_size=4,
    verbose=True
)
```
